# Load Dataset

```python
Df = pd.read_csv(r'C:\Users\cs\Desktop\Udemy_Project\vaibhav\Project_4\BitcoinPrice\BitcoinPrice.csv')
```

```python
Df.drop(['Date'], 1, inplace= True)
```

```python
Df.head()
#Now we have only Price Column
```

```python
#Predicting 30 days in Future
p_days = 30
Df['Prediction'] = Df[['Price']].shift(-p_days)
```

```python
Df.head()
#we have prediction value in starting of the dataset
```

```python
Df.tail()
#we don't have prediction value in last 30 days
```

```python
#Seperate x and y
X = np.array(Df.drop(['Prediction'], 1))
X = X[: len(Df) - p_days]
```

```python
X.shape
#pahle humaRE pass 365 days the & then humne last k 30 days hta diye to ab data me only 335 days hi remaining hai
```

```python
X
```

```python
#Creating Y as np Array
Y = np.array(Df['Prediction'])
Y = Y[: -p_days]
```

```python
Y.shape
```

```python
Y
```

```python
#Split the Dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
X_train.shape
```

```python
P_days_array = np.array(Df.drop(['Prediction'],  1))[-p_days:]
```

```python
print(P_days_array)
```

```python
from sklearn.ensemble import RandomForestRegressor 
```

# Random Forest Accuracy

```python
Rf = RandomForestRegressor(n_estimators= 1000, random_state= 1)
Rf.fit(X_train, Y_train)
print('Random Forest Accuracy: {:.2f}%'.format(Rf.score(X_test, Y_test)*100))
```

```python
#Prediction Value
Rf_predict = Rf.predict(X_test)
print(Rf_predict)
```

```python
#model value Predictrion for 30 days of the dataset
Rf_predict_30 = Rf.predict(P_days_array)
print(Rf_predict_30)
```
